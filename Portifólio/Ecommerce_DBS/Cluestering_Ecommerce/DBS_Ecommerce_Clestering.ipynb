Convertendo vari√°veis categoricas

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
db_ecommerce['Gender'] = encoder.fit_transform(db_ecommerce['Gender'])
db_ecommerce['Country'] = encoder.fit_transform(db_ecommerce['Country'])
db_ecommerce['State'] = encoder.fit_transform(db_ecommerce['State'])
db_ecommerce['Purchase Date'] = encoder.fit_transform(db_ecommerce['Purchase Date'])
db_ecommerce['Product Category'] = encoder.fit_transform(db_ecommerce['Product Category'])
db_ecommerce['Source'] = encoder.fit_transform(db_ecommerce['Source'])
```

```python
db_ecommerce.info()
```

```python
from scipy import stats
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputado = imputer.fit_transform(db_ecommerce)
X_imputado = pd.DataFrame(X_imputado, columns=db_ecommerce.columns)
print(X_imputado.isnull().sum())
```

```python
plt.scatter(db_ecommerce['Quantity'], db_ecommerce['State'])
plt.show()
```

```python
#cria o modelo de cluster KMeans com 3 clusters
est = KMeans(3)

#treina o modelo com os dados do Data Base
est.fit(db_ecommerce)

#faz a previs√£o dos cluters para cada ponto dos dados
y_kmeans = est.predict(db_ecommerce)

#criando um gr√°fico de dispers√£o, colorindo os pontos de acordo com o cluster ao qual pertecem
#plotando o gr√°fico de duas dimens√µes X e Y, com base nas duas primeiras colunas do Data Base
#c = y_kmeans defina a cor de cada ponto no gr√°fico
#s = 20 define o tamanho de cada ponto
#cmap = 'rainbow' define o mapa de cores, o rainbow √© uma paleta de cores com efeito de arco-iris
plt.scatter(db_ecommerce.iloc[:,0], db_ecommerce.iloc[:,1], c=y_kmeans, s=20, cmap='rainbow')

plt.xlabel("Texto1")
plt.ylabel("Texto2")

#est.cluster_centers_[:, 0] = acessa as coordenadas de X dos centr√≥ides
#est.cluster_centers_[:, 1] = acessa as coordenadas de Y dos centr√≥ides
#s = define o tamanho dos pontos dos centr√≥ides
#c = define a cor dos centroides neste caso preto
#label = define a legenda associada a esses pontos
plt.scatter(est.cluster_centers_[:, 0], est.cluster_centers_[:, 1], s = 50, c='black', label = 'Centroids')
```

```python
distorsions =[]
for k in range(2, 19):
          kmeans = KMeans(n_clusters=k)
          kmeans.fit(db_ecommerce)
          distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(10, 5))
plt.plot(range(2, 19), distorsions)
plt.xlabel('N√∫mero de clusters')
plt.ylabel('Soma dos quadrados do clueste')

plt.grid(True)
plt.title('Curva de cotovelo')
```

```python
from sklearn.preprocessing import Normalizer

values = Normalizer().fit_transform(db_ecommerce)
```

```python
kmeans = KMeans(n_clusters=4, n_init=10, max_iter=300)
y_pred = kmeans.fit_predict(values)
```

```python
from sklearn import metrics
labels = kmeans.labels_
```

```python
dbs = metrics.davies_bouldin_score(values, labels)
print(dbs)
```

```python
calinski = metrics.calinski_harabasz_score(values, labels)
print(calinski)
```
