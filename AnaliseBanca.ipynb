```python
import pandas as pd

df = pd.read_csv("Basebanca.csv")

#df = df.dropna()

# Remover colunas cujos nomes começam com 'Unnamed'
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]


display(df)
```

CRIAÇÃO DAS VARIÁVEIS

```python
# Filtrando o Data Frame com os meses
df_janeiro = df[df['Mes'] == 'Janeiro']
df_fevereiro = df[df['Mes'] == 'Fevereiro']
df_março = df[df['Mes'] == 'Março']
df_abril = df[df['Mes'] == 'Abril']
df_maio = df[df['Mes'] == 'Maio']
df_junho = df[df['Mes'] == 'Junho']
df_julho = df[df['Mes'] == 'Julho']
df_agosto = df[df['Mes'] == 'Agosto']
df_setembro = df[df['Mes'] == 'Setembro']
df_outubro = df[df['Mes'] == 'Outubro']
df_novembro = df[df['Mes'] == 'Novembro']
df_dezembro = df[df['Mes'] == 'Dezembro']
# Contar a frequência de cada produto vendido em cada mês
produtos_mais_vendidos_janeiro = df_janeiro['Produto'].value_counts()#nada
produtos_mais_vendidos_fevereiro = df_fevereiro['Produto'].value_counts()
produtos_mais_vendidos_março = df_março['Produto'].value_counts()#nada
produtos_mais_vendidos_abril = df_abril['Produto'].value_counts()
produtos_mais_vendidos_maio = df_maio['Produto'].value_counts()#nada
produtos_mais_vendidos_junho = df_junho['Produto'].value_counts()#nada
produtos_mais_vendidos_julho = df_julho['Produto'].value_counts()#nada
produtos_mais_vendidos_agosto = df_agosto['Produto'].value_counts()
produtos_mais_vendidos_setembro = df_setembro['Produto'].value_counts()
produtos_mais_vendidos_outubro = df_outubro['Produto'].value_counts()
produtos_mais_vendidos_novembro = df_novembro['Produto'].value_counts()
produtos_mais_vendidos_dezembro = df_dezembro['Produto'].value_counts()#nada

print(produtos_mais_vendidos_janeiro)
```

```python
import matplotlib.pyplot as plt

# Gerar o gráfico de barras
produtos_mais_vendidos_novembro.plot(kind='bar', color='skyblue')

# Personalizar o gráfico
plt.title('Produtos Mais Vendidos em Novembro')
plt.xlabel('Nome dos Produtos')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta o layout para evitar cortes

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gerar o gráfico de barras
produtos_mais_vendidos_fevereiro.plot(kind='bar', color='skyblue')

# Personalizar o gráfico
plt.title('Produtos Mais Vendidos em Fevereiro')
plt.xlabel('Nome dos Produtos')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta o layout para evitar cortes

# Mostrar o gráfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


# Converter para DataFrame para usar com Seaborn
produtos_mais_vendidos_setembro = produtos_mais_vendidos_setembro.reset_index()
produtos_mais_vendidos_setembro.columns = ['Produto', 'Quantidade']

# Gerar o gráfico de barras
sns.barplot(data=produtos_mais_vendidos_setembro, x='Produto', y='Quantidade', palette='Blues_d')

# Personalizar o gráfico
plt.title('Produtos Mais Vendidos em Janeiro')
plt.xlabel('Nome dos Produtos')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```
