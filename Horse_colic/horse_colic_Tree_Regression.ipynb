Processamento dos dados

```python
#IMPORTANDO AS BIBLIOTECAS QUE SERÃO USADAS
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Criando a variavel df, contendo a base de dados de cavalos
data = pd.read_csv("horse.csv")

#renomeando o  nome das colunas do dfbase
data.columns = ['Surgery','Age','Hospital Number','Rectal Temperature','Pulse','Respiratory Rate','Temperature of Extremities','Peripheral Pulse','Mucous Membranes','Capillary Refill Time','Pain','Peristalsis','Abdominal Distension','Nasogastric Tube','Nasogastric Reflux','Nasogastric Reflux PH','Rectal Examination Feces','Abdomen','Packed Cell Volume','Total Protein','Abdominocentesis Appearance','Abdominocentesis Total Protein','Outcome','Surgical Lesion','Lesion Site','Lesion Type','Lesion Subtype','CP df']

#Mostra as 5 primeiras linhas do dfFrame
data.head() # Show first 5 samples

```

```python
df = data.copy()
```

Tratamento de Variáveis

```python
# Importa o LabelEncoder, usado para converter variáveis categóricas em números
from sklearn.preprocessing import LabelEncoder
# Cria o objeto encoder que fará a transformação das variáveis categóricas
encoder = LabelEncoder()
# Aplica o encoder na coluna 'Surgery', transformando categorias em valores numéricos
df['Surgery'] = encoder.fit_transform(df['Surgery'])
df['Age'] = encoder.fit_transform(df['Age'])
df['Temperature of Extremities'] = encoder.fit_transform(df['Temperature of Extremities'])
df['Peripheral Pulse'] = encoder.fit_transform(df['Peripheral Pulse'])
df['Mucous Membranes'] = encoder.fit_transform(df['Mucous Membranes'])
df['Capillary Refill Time'] = encoder.fit_transform(df['Capillary Refill Time'])
df['Pain'] = encoder.fit_transform(df['Pain'])
df['Peristalsis'] = encoder.fit_transform(df['Peristalsis'])
df['Abdominal Distension'] = encoder.fit_transform(df['Abdominal Distension'])#
df['Nasogastric Tube'] = encoder.fit_transform(df['Nasogastric Tube'])#
df['Nasogastric Reflux'] = encoder.fit_transform(df['Nasogastric Reflux'])
df['Rectal Examination Feces'] = encoder.fit_transform(df['Rectal Examination Feces'])
df['Abdomen'] = encoder.fit_transform(df['Abdomen'])
df['Abdominocentesis Appearance'] = encoder.fit_transform(df['Abdominocentesis Appearance'])
df['Outcome'] = encoder.fit_transform(df['Outcome'])
df['Surgical Lesion'] = encoder.fit_transform(df['Surgical Lesion'])
df['CP df'] = encoder.fit_transform(df['CP df'])

#df.head()
```

Validando se existem dados faltantes

```python
#df.isnull().sum()
```

Substituindo os dados faltantes pela média

```python
from sklearn.impute import SimpleImputer
```

```python
# Cria o imputador que preenche NaN com a média dos dados
imputer = SimpleImputer(strategy='mean')
# Aplica imputação
X_imputado = imputer.fit_transform(df)
X_imputado = pd.DataFrame(X_imputado, columns=df.columns)
#verifica os NAN
#print(X_imputado.isnull().sum())

#df = X_imputado

```

Construindo Árvore de Decisão


```python
# Importa a função train_test_split para dividir os dados em treino e teste
from sklearn.model_selection import train_test_split

# Define as variáveis independentes (todas as colunas exceto 'Surgery')
X = df.drop('Surgery', axis=1)

# Define a variável dependente (o alvo a ser previsto)
y = df['Surgery']

# Divide os dados em conjuntos de treino e teste:
# - 20% dos dados vão para teste
# - random_state=45 garante que a divisão seja reproduzível
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, random_state=45)
```

```python
print(len(X_treino))
print(len(X_teste))
```

```python
# Importa o DecisionTreeRegressor, usado para regressão com árvore de decisão
from sklearn.tree import DecisionTreeRegressor

# Cria o modelo de árvore de decisão para regressão, definindo uma semente para reprodutibilidade
dtr = DecisionTreeRegressor(random_state=45)

# Treina o modelo nos dados de treino
dtr.fit(X_treino, y_treino)
```

Aplicando Métricas para validar o modelo

```python
from sklearn import metrics
```

```python
print(f"Número de nós: {dtr.tree_.node_count}")
print(f"Número de folhas: {dtr.tree_.n_leaves}")
print(f"Profundidade máxima: {dtr.tree_.max_depth}")
```

R² é usado para medir o quanto o modelo explica a variablidade dos dados

O MAE mostra o erro médio

RMSE é o erro média quadrado

```python
print('Métricas conjunto de treino:')
print('R²:', metrics.r2_score(y_treino, dtr.predict(X_treino)))
print('MAE:', metrics.mean_absolute_error(y_treino, dtr.predict(X_treino)))
print('RMSE:', metrics.mean_squared_error(y_treino, dtr.predict(X_treino)))

print('\n\nMétricas conjunto de teste:')
print('R²:', metrics.r2_score(y_teste, dtr.predict(X_teste)))
print('MAE:', metrics.mean_absolute_error(y_teste, dtr.predict(X_teste)))
print('RMSE:', metrics.mean_squared_error(y_teste, dtr.predict(X_teste)))
```

Resolvendo problema de Overfiting

```python
param_grid = {
    'max_depth': [2, 4, 6, 8], # Profundidade Máxima da árvore
    'min_samples_leaf': [2, 5, 10], #Número mínimo de amostras para estar em um nó
    'min_samples_split': [5, 10, 15], #Mínimo de amostras que um nó deve ter antes de ser divido
    'max_leaf_nodes': [None, 10, 20] #Número máximo de nós da folha da arvore
}
```

```python
# Importa o GridSearchCV, usado para buscar automaticamente os melhores hiperparâmetros de um modelo
from sklearn.model_selection import GridSearchCV

# Cria o objeto GridSearchCV com:
# - modelo dtr
# - grade de parâmetros param_grid
# - métrica de avaliação RMSE negativo
# - validação cruzada com 5 folds
# - n_jobs=-1 para usar todos os núcleos do processador
grid_search = GridSearchCV(dtr, param_grid, scoring='neg_root_mean_squared_error', cv=5, n_jobs=-1)

# Treina o GridSearchCV nos dados de treino, testando todas as combinações de parâmetros
grid_search.fit(X_treino, y_treino)

# Recupera o modelo otimizado com os melhores parâmetros encontrados pelo GridSearch
dtr_otimizado = grid_search.best_estimator_

# Exibe os melhores parâmetros encontrados na busca
print('Melhores parâmetros:', grid_search.best_params_)
```

```python
print('Métricas conjunto de treino:')
print('R²:', metrics.r2_score(y_treino, dtr_otimizado.predict(X_treino)))
print('MAE:', metrics.mean_absolute_error(y_treino, dtr_otimizado.predict(X_treino)))
print('RMSE:', metrics.mean_squared_error(y_treino, dtr_otimizado.predict(X_treino)))

print('\n\nMétricas conjunto de teste:')
print('R²:', metrics.r2_score(y_teste, dtr_otimizado.predict(X_teste)))
print('MAE:', metrics.mean_absolute_error(y_teste, dtr_otimizado.predict(X_teste)))
print('RMSE:', metrics.mean_squared_error(y_teste, dtr_otimizado.predict(X_teste)))
```

Uilizando validação Cruzada

```python
from sklearn.model_selection import cross_validate, KFold

cv_estrategia = KFold(n_splits=5, shuffle=True, random_state=45)

resultados = cross_validate(dtr_otimizado,X,y,scoring='neg_root_mean_squared_error', cv= cv_estrategia,
                            return_train_score=True)
```

```python
treino_rmse = np.sqrt(-resultados['train_score'])
teste_rmse = np.sqrt(-resultados['test_score'])


print("Treino RMSE em cada fold:",treino_rmse)
print("Treino RMSE em cada fold:",teste_rmse)
print("\nMédia RMSE em cada fold:",treino_rmse.mean())
print("Média RMSE em cada fold:",treino_rmse.mean())
```
